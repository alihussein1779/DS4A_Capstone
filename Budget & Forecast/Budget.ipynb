```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
"""

    Calculate the breakdown of costs depending on capital
    
    Arguments:
    Different values calculated from the capital
    
    Output:
    answer: Pie Chart, results of above calculations


"""

capital = int(input('How much is your budget?\n'))


def breakdown_of_costs():
    
 
     
    """"
    These are the different  costs that should be expected. 
    Some of these costs such as energy and water will be recurring but this helps to break down initial costs.
    They are obtained by multiplying the expected percentage out of 100 by the capital to get the 
        expected cost from that capital
    """
    
    avg_operational_cost= capital * 0.56
    avg_daily_water_spending =  capital * 0.114
    avg_monthly_energy_usage_2 = capital * 0.216
    avg_cost_of_growing_material = capital * 0.11
    avg_monthly_general_spending = 200
    
    
    """This puts the information above into a list so that we can plot it as a pie chart"""
    group = [avg_operational_cost,   avg_monthly_energy_usage_2, avg_cost_of_growing_material,  avg_daily_water_spending]
    total = avg_operational_cost +  avg_monthly_energy_usage_2 + avg_cost_of_growing_material +  avg_daily_water_spending
    LBS = ["Operational Costs", "Monthly Energy Usage", "Cost of Growing Material", "Monthly Water Expenditure"]
    
    fig = plt.figure(figsize =(10, 7))
    plt.pie(group, labels = LBS)
    
    #return fig
    return "Your total average operations cost in the beginning is ", avg_operational_cost, 
    "Your total average energy cost in the beginning is ",  avg_monthly_energy_usage_2,"Your total average growing material cost in the beginning is ", avg_cost_of_growing_material, "Your total average water cost in the beginning is ", avg_daily_water_spending

breakdown_of_costs()
```
